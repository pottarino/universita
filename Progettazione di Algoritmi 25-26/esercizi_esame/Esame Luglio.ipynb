```python
def esercizio1(M, s):
    vettore_distanze = [0 for i in range(len(M))] #O(n)
    visitati = [False for i in range(len(M))] #O(n)
    coda = [s] #O(1)
    index = 0 #O(1)
    res = [0 for i in range(len(M))] #O(n)
    res[0] = 1 #O(1)
    visitati[s] = True
    while index < len(coda): #la coda è lunga al massimo n
        elemento = coda[index]
        for j in range(len(M)): #O(n)
            if M[elemento][j]:
                if not visitati[j]:
                    coda.append(j)
                    vettore_distanze[j] = vettore_distanze[elemento] + 1
                    visitati[j] = True
                    res[vettore_distanze[j]] += 1
        index += 1
    return res
#la complessità di questo algoritmo è O(n^2) per un grafo completo
```
