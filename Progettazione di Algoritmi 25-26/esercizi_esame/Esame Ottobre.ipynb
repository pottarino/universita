```python
def esercizio1(G, x):
    stack = [0]
    visitati = [False for _ in range(len(G))]
    visitati[0] = True
    while stack:
        item = stack.pop()
        if item == x:
            return True
        for vicino in G[item]:
            if vicino % 2 == 0 and not visitati[vicino]:
                visitati[vicino] = True
                stack.append(vicino)
    return False

```

```python
def esercizio2(A):
    n = len(A)
    if n == 0:
        return 0

    # T[i][0] = numero di sottosequenze pari che terminano in A[i]
    # T[i][1] = numero di sottosequenze dispari che terminano in A[i]
    T = [[0, 0] for _ in range(n)]

    # primo elemento
    T[0][A[0] % 2] = 1

    for i in range(1, n):
        parity = A[i] % 2
        # sottosequenza solo [A[i]]
        T[i][parity] = 1

        # aggiungere A[i] a sottosequenze non adiacenti precedenti
        if i >= 2:
            for p in [0,1]:
                new_parity = (p + parity) % 2
                T[i][new_parity] += T[i-2][p]

    # risultato: somma di tutte le sottosequenze pari
    return sum(T[i][0] for i in range(n))

print(esercizio2([1,2,5,4,6]))
```

```python
def esercizio3(n, sol=None):
    if sol is None:
        sol = []
    i = len(sol)
    if len(sol) == n:
        print("".join(sol))
        return
    posso_zero = not ( (i == 1 and sol[0] == "1") or i > 1 and sol[-1] == "1" and sol[-2] == "0")
    if posso_zero:
        sol.append("0")
        esercizio3(n, sol)
        sol.pop()

    posso_uno = not (n - i - 1  == 0 and (i==0 or sol[-1] == "0"))

    if posso_uno:
        sol.append("1")
        esercizio3(n, sol)
        sol.pop()



esercizio3(3)
```
