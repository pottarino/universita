```python
def esercizio1(G):
    visitati = [False for _ in range(len(G))]
    amici = [1 for _ in range(len(G))]
    for nodo in range(len(G)):
        if not visitati[nodo]:
            visitati[nodo] = True
            vicini = [nodo]
            da_visitare = [nodo]
            while da_visitare:
                nodo = da_visitare.pop()
                for vicino in G[nodo]:
                    if not visitati[vicino]:
                        visitati[vicino] = True
                        vicini.append(vicino)
            n_vicini = len(vicini)
            for vicino in vicini:
                amici[vicino] = n_vicini
```

```python
def esercizio2(A):
    n = len(A)
    if not A:
        return 0
    if len(A) <= 2:
        return sum(A)
    T = [A[0], A[0] + A[1]]
    for i in range(2, n):
        T.append(A[i] + max(T[i-1], T[i-2]))
    return T[-1]

print(esercizio2([1,4,2,3,-1,-5,1]))
```

```python
def esercizio3(n):
    esercizio3r(n, 0, [])

def esercizio3r(n, i, sol):
    if len(sol) == n:
        print("".join(sol))
        return
    alfabeto = {"0","1","2"}
    for item in alfabeto:
        if i < 2 or int(item) + int(sol[i-1]) + int(sol[i-2]) in {4,5}:
            sol.append(item)
            esercizio3r(n, i + 1, sol)
            sol.pop()

esercizio3(5)
```
