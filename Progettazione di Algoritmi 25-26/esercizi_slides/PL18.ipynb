```python
def stringhe_numeriche(n, sol = None, messi = None):
    if sol is None:
        sol = []
    if messi is None:
        messi = [False for i in range(n+1)]
    if len(sol) == n:
        print("".join(sol))
        return
    for i in range(1, n+1):
        if len(sol) < 1 or int(sol[-1]) % 2 != i % 2 and not messi[i]:
            sol.append(str(i))
            messi[i] = True
            stringhe_numeriche(n, sol, messi)
            sol.pop()
            messi[i] = False

stringhe_numeriche(4)

```

```python
def stringhe_ternarie(n, sol = None, n_a = 0, n_b = 0, n_c = 0):
    if sol is None:
        sol = []
    if len(sol) == n:
        print("".join(sol))
        return
    sol.append("b")
    stringhe_ternarie(n, sol, n_a, n_b + 1, n_c)
    sol.pop()
    rem = n - len(sol) - 1
    n_a2 = n_a + 1
    if rem > n_a2 - n_b + 1:
        sol.append("a")
        stringhe_ternarie(n, sol, n_a2, n_b, n_c)
        sol.pop()
    n_c2 = n_c + 1
    if rem > n_c2 - n_b + 1:
        sol.append("c")
        stringhe_ternarie(n, sol, n_a, n_b, n_c2)
        sol.pop()

stringhe_ternarie(5)
```

```python
def scs(X, i = 0, sol = None, last = -1):
    if sol is None:
        sol = []
    if len(sol) == len(X):
        print("".join(sol))
        return
    if last < int(X[i]):
        sol.append(str(X[i]))
        scs(X, i + 1, sol, int(X[i]))
        sol.pop()
    sol.append("*")
    scs(X, i + 1, sol, last)
    sol.pop()

scs("0120")

```

```python
def sdm(X, i = 0, sol = None):
    if sol is None:
        sol = []
    if len(sol) == len(X):
        print("".join(sol))
        return
    for alpha in "012":
        if alpha != X[i]:
            sol.append(alpha)
            sdm(X, i + 1, sol)
            sol.pop()

sdm("02012")
```
