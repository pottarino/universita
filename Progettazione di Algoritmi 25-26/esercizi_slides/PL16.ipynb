```python
def dadi(n, s, f):
    T = [[None for _ in range(s+1)] for _ in range(n+1)]
    return dadir(n, s, f, T)

def dadir(i,j,f,T):
    if T[i][j] is not None:
        return T[i][j]
    elif i == 0 and j == 0:
        T[i][j] = 1
    elif i == 0 or j == 0:
        T[i][j] = 0
    else:
        T[i][j] = 0
        for k in range(1, f+1):
            if k <= j:
                T[i][j] += dadir(i-1, j - k, f, T)
    return T[i][j]

print(dadi(2,3,6))
```

```python
def stringhe_senza_11(n):
    T = [1, 3]
    for i in range(2, n+1):
        T.append(2*(T[i-1] + T[i-2]))
    return T[n]

print(stringhe_senza_11(4))
```

```python
def carte(C):
    N = len(C)
    if N == 0:
        return 0
    T = [[None for _ in range(N + 1)] for _ in range(N + 1)]
    return carteR(C, 0, N, T)


def carteR(C, i, j, T):
    if T[i][j] is not None:
        return T[i][j]
    if i >= j:
        return 0
    if j - i == 1:
        return C[i]
    if j - i == 2:
        return max(C[i], C[j - 1])

    T[i][j] = max(
        C[i] + min(carteR(C, i + 2, j, T), carteR(C, i + 1, j - 1, T)),
        C[j - 1] + min(carteR(C, i, j - 2, T), carteR(C, i + 1, j - 1, T))
    )

    return T[i][j]
print(carte([2, 10, 5, 1]))  # Output: 11
```

```python
    combinazioni = ["abc", "acb", "cba", "cab", "bca", "bac", "bab", "bcb", "cac", "cbc", "aba", "aca"]

def compatibili(conf1, conf2):
    return conf1[0] != conf2[0] and conf1[1] != conf2[1] and conf1[2] != conf2[2]


def matrice_3xn(n):
    if n == 0:
        return 1
    T = [[None for _ in range(12)] for _ in range(n)]
    totale_modi = 0
    for c in range(12):
        totale_modi += conta_colorazioni_R(n - 1, c, T)

    return totale_modi


def conta_colorazioni_R(i, conf_corrente, T):
    if T[i][conf_corrente] is not None:
        return T[i][conf_corrente]
    if i == 0:
        return 1

    modi = 0
    for conf_precedente in range(12):
        stringa_corrente = combinazioni[conf_corrente]
        stringa_precedente = combinazioni[conf_precedente]

        if compatibili(stringa_corrente, stringa_precedente):
            modi += conta_colorazioni_R(i - 1, conf_precedente, T)

    T[i][conf_corrente] = modi
    return T[i][conf_corrente]

print(matrice_3xn(1))
```
