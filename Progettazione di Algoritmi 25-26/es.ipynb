```python
def es(G, s, t):
    T = [None] * len(G)
    res = traccia(grafoTrasposto(G), s, t, T)
    return res

def traccia(G, s, t, T):
    if T[t] is not None:
        return T[t]
    if t == s:
        T[t] = 1
    else:
        T[t] = 0
        for vicino in G[t]:
            T[t] += traccia(G,s,vicino, T)
    return T[t]

def grafoTrasposto(G):
    GT = [[] for i in range(len(G))]
    for u in range(len(G)):
        for v in G[u]:
            GT[v].append(u)  # Inverte l'arco: da u->v a v->u
    return GT


import Algoritmi2 as a

g = a.grafoLista([[1,2,3,4,5,6,7],
                  [2,3,4,5,6,7],
                  [3,4,5,6,7],
                  [4,5,6,7],
                  [5,6,7],
                  [6,7],
                  [7],
                  []
                  ], oriented=True)

g.plot()
print(es(g, 3, 5))
```

```python
def esercizio(s):
    resti = [0, 0, 0]
    if len(s) == 0:
        return 0
    resti[int(s[0]) % 3] = 1
    sol = resti[0]
    for carattere in s[1:]:
        n = int(carattere)
        r = n % 3
        if r == 0:
            resti[0] += 1
        elif r == 1:
            resti = resti[1], resti[2] + 1, resti[0]
        else:
            resti = resti[2], resti[0], resti[1] + 1
        sol += resti[0]
    return sol

print(esercizio("1111"))
```
