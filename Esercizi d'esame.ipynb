##  \$[Grafi]\$
Dato un grafo diretto \$G\$ con \$n\$ nodi e \$m\$ archi memorizzato tramite liste di adiacenza, tre suoi vertici \$A\$, \$B\$ e \$C\$, e un intero \$K\$,

progettare un algoritmo che, in tempo
\begin{equation}O(n + m)\end{equation} restituisca \$True\$ se esiste
un cammino con al più \$k\$ archi da \$A\$ a \$C\$ che passa per \$B\$, \$False\$ altrimenti.

```python
def esercizio1(G, a, b, c, k):
    G.plot()
    v1 = vettore_distanze(G, a)
    v2 = vettore_distanze(G, b)
    return v1[b] + v2[c] <= k

from collections import deque

def vettore_distanze(G, a):
    vett = [float("inf") for el in range(len(G))]
    coda = deque()
    coda.append(a)
    vett[a] = 0
    while coda:
        el = coda.popleft()
        for vicino in G[el]:
            if vett[vicino] == float("inf"):
                coda.append(vicino)
                vett[vicino] = vett[el] + 1
    return vett
```

##  [Conteggio]
Si considerino le stringhe binarie di lunghezza \$n\$ che non contengono tre 1 consecutivi.
Progettare un algoritmo di programmazione dinamica che, in tempo
\begin{equation}O(n)\end{equation}
calcoli il numero totale di tali stringhe.

```python
def esercizio_2_supp(n):
    '''
    in T[i] ci va una stringa che non termini con 111
    se T[i] termina con 0, T[i-1]
    altrimenti se T[i] termina con 1:
        se T[i-1] termina con 0, T[i-2]
        se T[i-1] termina con 1,
            se T[i-2] termina con 0, T[i-3]
            se T[i-3] termina con 1, 0
    '''
    T = [None for _ in range(n+1)]
    T[0] = 1
    T[1] = 2
    T[2] = 4
    for i in range(3, n+1):
        T[i] = T[i-1] + T[i-2] + T[i-3]
    return T

def esercizio_2(n):
    return esercizio_2_supp(n)[n]
```
